```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None


# Import the csv file
df = pd.read_csv(r'C:\Users\DELL\Desktop\Python\movies.csv\movies.csv')
```

```python
df.head()
```

```python
# the sum of the null values preset in the dataset column-wise

df.isnull().sum()
```

```python
# To determine that there is no missing value

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Dropping Rows with Missing Data in the DataFreame

df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
# To Check if missing Data has been dropped

df = df.dropna()

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# To check the data types

print(df.dtypes)
```

```python
# To drop duplicates

df.drop_duplicates()
```

```python
sns.regplot(x='gross', y='budget', data=df, color='green', marker='+')
```

```python
# Colleration matrix between column

df.corr(method = 'pearson')

# Matrices with >5 shows a high level of correlation
```

```python
# Colleration matrix between column

df.corr(method = 'kendall')

# Matrices with >5 shows a high level of correlation
```

```python
# Colleration matrix between column

df.corr(method = 'spearman')

# Matrices with >5 shows a high level of correlation
```

```python
corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot = True)

plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

```

```python
corr_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind='quicksort')

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized        
        
```

```python
df_numerized.corr(method='pearson')
```

```python
df_numerized_matrix = df_numerized.corr(method='pearson')

sns.heatmap(df_numerized_matrix, annot = True)

plt.title("Correlation Matrix for Movies")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```
